```python
# Import Libraries for the project
from sklearn import tree  # Import Decision Tree Classifier as tree
from sklearn.neighbors import KNeighborsClassifier  # Import for K-Nearest Classifier
from sklearn.naive_bayes import GaussianNB  # Import for Naive Bayes
from sklearn import svm # Import for Support Vector Machine
from sklearn.neural_network import MLPClassifier # Import for multilayer perceptron
import pandas as pd # Import pandas
import numpy as np # Import numpy
import matplotlib.pyplot as plt # For data visualization
from sklearn.model_selection import train_test_split  # for train_test_split function
from sklearn.metrics import accuracy_score  #  for accuracy calculation
from sklearn.metrics import cohen_kappa_score # for kappa calculation
```

```python
data = pd.read_csv("CE802_P2_Data/CE802_P2_Data.csv") # Imported the dataset using pandas
```

```python
data # FIrst look at the data
```

```python
# Approach without F15 column as it has NaN 
train_data_x = data.iloc[:,:14] # splitted data for the features
train_data_y = data.iloc[:, 15] # Splitted for the target column Class
x_train, x_test, y_train, y_test = train_test_split(train_data_x, train_data_y, test_size = 0.2, random_state = 10) # created the test and train data
```

```python
#creating a Decision Tree classifier object
DTCO = tree.DecisionTreeClassifier(criterion = 'gini', random_state = 1)
DTCO = DTCO.fit(x_train, y_train) # Training the dataset
DTpredict = DTCO.predict(x_test) # Predict the response for the test data
```

```python
#Accuracy and Kappa fro DT
DTacc1 = accuracy_score(y_test, DTpredict)
DTkappa1 = cohen_kappa_score(y_test, DTpredict)
print("Accuracy:", DTacc1)
print("Kappa Statistics:", DTkappa1)
```

```python
#creating a K-Nearest classifier object
KNCO = KNeighborsClassifier(n_neighbors = 7)
KNCO = KNCO.fit(x_train, y_train) # Training the dataset
KNpredict = KNCO.predict(x_test) # Predict the response for the test data
```

```python
#Accuracy and Kappa fro KN
KNacc1 = accuracy_score(y_test, KNpredict)
KNkappa1 = cohen_kappa_score(y_test, KNpredict)
print("Accuracy:", KNacc1)
print("Kappa Statistics:", KNkappa1)
```

```python
#creating a Naive Bayes classifier object
NBCO = GaussianNB()
NBCO = NBCO.fit(x_train, y_train) # Training the dataset
NBpredict = NBCO.predict(x_test) # Predict the response for the test data
```

```python
#Accuracy and Kappa fro NB
NBacc1 = accuracy_score(y_test, NBpredict)
NBkappa1 = cohen_kappa_score(y_test, NBpredict)
print("Accuracy:", NBacc1)
print("Kappa Statistics:", NBkappa1)
```

```python
#creating a Support Vector Machines classifier object
SVCO = svm.SVC() 
SVCO = SVCO.fit(x_train, y_train) # Training the dataset
SVpredict = SVCO.predict(x_test) # Predict the response for the test data
```

```python
#Accuracy and Kappa fro SVN
SVacc1 = accuracy_score(y_test, SVpredict)
SVkappa1 = cohen_kappa_score(y_test, SVpredict)
print("Accuracy:", SVacc1)
print("Kappa Statistics:", SVkappa1)
```

```python
#creating a Multi-layer Perceptron classifier object
MPCO = MLPClassifier(hidden_layer_sizes = (4,), random_state = 1) 
MPCO = MPCO.fit(x_train, y_train) # Training the dataset
MPpredict = MPCO.predict(x_test) # Predict the response for the test data
```

```python
#Accuracy and Kappa fro MP
MPacc1 = accuracy_score(y_test, MPpredict)
MPkappa1 = cohen_kappa_score(y_test, MPpredict)
print("Accuracy:", MPacc1)
print("Kappa Statistics:", MPkappa1)
```
